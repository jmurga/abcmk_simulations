```python
import sys
import pandas as pd
import numpy as np
import glob
from multiprocessing import Pool
from plotnine import *
import plotnine as plt

PATH = "/home/jmurga/mkt/202004" 
sys.path.insert(0, PATH + '/scripts/src/')  
from pyAmkt import *
from slimParser import *
```

```python
%load_ext rpy2.ipython
```

```python
def parsingTwoepochs(table,equilibrium,time,l,h,case,factors):
    
    sEq = pd.read_csv(equilibrium + "/sfs.tsv",sep='\t')
    dEq = pd.read_csv(equilibrium + "/div.tsv",sep='\t')
    cEq = cumulativeSfs(sEq.to_numpy())

    aEq1 = amkt(cEq[:,[0,4,2]],dEq.to_numpy().flatten(),l,h)
    aEq2 = amkt(cEq[:,[0,1,2]],dEq.to_numpy().flatten(),l,h)
    f = np.arange(1,aEq1[1].shape[0]+1)/cEq.shape[0]

    
    eqAlpha = pd.DataFrame({'f':f,'alphas':aEq1[1],'B':np.repeat(0.999,f.shape[0]),'alphaW':np.repeat(0.1,f.shape[0]),'sfs':np.repeat(['All alleles'],f.shape[0]),'analysis':'Equilibrium','time':str(time)})
    eqAlpha = pd.concat([eqAlpha,pd.DataFrame({'f':f,'alphas':aEq2[1],'B':np.repeat(0.999,f.shape[0]),'alphaW':np.repeat(0.1,f.shape[0]),'sfs':np.repeat(['Neutral + deleterious'],f.shape[0]),'analysis':'Equilibrium','time':str(time)})])
    tmpEq = pd.DataFrame({'trueAlpha':(dEq.ds+dEq.dw)/dEq.di,'asymp_nopos':aEq1[0]['alpha'],'asymp':aEq2[0]['alpha'],'analysis':'Equilibrium','time':str(time)})

        
    out = [];al = [];cumulative=True
    for i in table:
        sfs = pd.read_csv(i + "/sfs.tsv",header=0,sep="\t")
        div = pd.read_csv(i + "/div.tsv",header=0,sep="\t")

        trueAlpha = (div.ds+div.dw)/div.di
        print(i)
        if cumulative:
            cSfs = cumulativeSfs(sfs.to_numpy())
        else:
            cSfs = sfs.to_numpy()

        asymp1 = amkt(cSfs[:,[0,4,2]],div.to_numpy().flatten(),l,h)

        asymp2 = amkt(cSfs[:,[0,1,2]],div.to_numpy().flatten(),l,h)	
        f = np.arange(1,asymp1[1].shape[0]+1)/cSfs.shape[0]

        tmpAlpha = pd.DataFrame({'f':f,'alphas':asymp1[1],'B':np.repeat(0.999,f.shape[0]),'alphaW':np.repeat(0.1,f.shape[0]),'sfs':np.repeat(['All alleles'],f.shape[0]),'analysis':i.split('/')[-1],'time':str(time)})

        tmpAlpha = pd.concat([tmpAlpha,pd.DataFrame({'f':f,'alphas':asymp2[1],'B':np.repeat(0.999,f.shape[0]),'alphaW':np.repeat(0.1,f.shape[0]),'sfs':np.repeat(['Neutral + deleterious'],f.shape[0]),'analysis':i.split('/')[-1],'time':str(time)})])
        tmp = pd.DataFrame({'trueAlpha':trueAlpha,'asymp_nopos':asymp1[0]['alpha'],'asymp':asymp2[0]['alpha'],'analysis':i.split('/')[-1],'time':str(time)})
        out.append(tmp)
        al.append(tmpAlpha)
    
    dfAlpha=pd.concat(al)
    dfAlpha = pd.concat([dfAlpha,eqAlpha])
    
    dfAlpha.analysis = dfAlpha.analysis.str.replace('twoepochs_s','')
    dfAlpha.analysis = dfAlpha.analysis.str.replace('1',case + ' strength = ' + str(factors[0]))
    dfAlpha.analysis = dfAlpha.analysis.str.replace('2',case + ' strength = ' + str(factors[1]))
    dfAlpha.analysis = dfAlpha.analysis.str.replace('3',case + ' strength = ' + str(factors[2]))
    dfAlpha.analysis = dfAlpha.analysis.str.replace('4',case + ' strength = ' + str(factors[3]))
    
    out = pd.concat(out)
    out = pd.concat([out,tmpEq])
    
    return(out,dfAlpha)
```
