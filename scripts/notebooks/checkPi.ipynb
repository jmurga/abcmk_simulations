```python
import allel
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
callset = allel.read_vcf("/home/jmurga/mkt/202004/results/noBgsSingleLocus.vcf")
```

```python
pos = callset['variants/POS']
gt = allel.GenotypeArray(callset['calldata/GT'])
ac = gt.count_alleles()
```

```python
pi = allel.mean_pairwise_difference(ac)
```

```python
pi, windows, n_bases, counts = allel.windowed_diversity(pos, ac, size=10000, step=1000,start=1, stop=pos[-1])
```

```python
dfPi = pd.DataFrame({'pi':pi,'positions':windows[:,1]});dfPi
```

```python
sns.lineplot(x='positions',y='pi',data=dfPi)
```
