```julia
"""
Executing ABC-MK and Grapes if required
"""
function parse_inference(param,folder;h5_file::String,grapes::Bool=false)

    # mkpath(output);
    analysis = split(folder,"/")[end]

    divergence = CSV.read.(filter(x->occursin("div",x),readdir(folder,join=true)),DataFrame,header=true);

    alpha = vcat(divergence...)
    alpha = DataFrame([alpha[:,3]./alpha[:,1],alpha[:,4]./alpha[:,1],sum.(eachrow(alpha[:,3:4]))./alpha[:,1],fill(analysis,length(divergence)),fill("slim",length(divergence))],[:α_weak,:α_strong,:α,:analysis,:method])
    
    divergence = map(x->Int64.(Array(x)[:,1:2]),divergence)

    sfs = CSV.read.(filter(x->occursin("sfs",x),readdir(folder,join=true)),DataFrame,header=true);
    sfs = map(x -> Array(x)[:,1:3],sfs)
    map(x -> x[:,1] .= 1:(param.nn-1),sfs)

    out = @suppress begin
        summstat = MKtest.summary_statistics(param,sfs=sfs,divergence=divergence,h5_file=h5_file,analysis_folder= folder,summstat_size=10^5);
        posteriors = MKtest.ABCreg(analysis_folder=folder,P=5,S=length(param.dac),tol=0.025,abcreg="/home/jmurga/ABCreg/src/reg");
        out = MKtest.summary_abc(posteriors);
    end

    insertcols!(out[2],:analysis=>analysis);
    insertcols!(out[2],:method=>"abc_mk");

    if grapes
        m = fill(Int64.(25000*2000 .* [0.75 0.25]),length(sfs))
        df_gammazero = MKtest.grapes(sfs[1:10], divergence[1:10], m[1:10], "GammaZero", folder, 20)
        df_gammazero =  df_gammazero[:,["model","alpha","GammaZero:negGmean","GammaZero:negGshape"]]
        insertcols!(df_gammazero,1,:analysis=>analysis);
        rename!(df_gammazero,[:analysis,:model,:alpha,:gamma,:shape])
        
        df_gammaexpo = MKtest.grapes(sfs, divergence, m, "GammaExpo", folder, 20)
        df_gammaexpo =  df_gammaexpo[:,["model","alpha","GammaExpo:negGmean","GammaExpo:negGshape"]]
        insertcols!(df_gammaexpo,1,:analysis=>analysis);
        rename!(df_gammaexpo,[:analysis,:model,:alpha,:gamma,:shape])
        
        df_grapes = vcat(df_gammazero,df_gammaexpo)
        
        return([alpha,out[2]],df_grapes)
    else
        return([alpha,out[2]])
    end

end
```

# Two-epoch model

Because we simulated $\alpha = 0.2$, we will re-estimate the rates using a prior of $\alpha = [0.1,0.5]$ instead of the default $\alpha = [0.1,0.9]$ and a two differents prior of BGS values.

```julia
adap = MKtest.parameters(n=50,al=0.184,be=0.184/457,dac=[1,2,4,5,10,15,20,25,30,40,50,70],B_bins=collect(0.3:0.025:0.5));
@time df = MKtest.rates(param = adap,alpha=[0.1,0.5],gH=[200,2000],gL=[1,10],gam_dfe=[-800,-200],gam_flanking=[-1000,-500],output= path * "/raw_data/rates_twoepoch_0.4.jld2");
adap = MKtest.parameters(n=50,al=0.184,be=0.184/457,dac=[1,2,4,5,10,15,20,25,30,40,50,70],B_bins=push!(collect(0.8:0.025:0.975),0.999));
@time df = MKtest.rates(param = adap,alpha=[0.1,0.5],gH=[200,2000],gL=[1,10],gam_dfe=[-800,-200],gam_flanking=[-1000,-500],iterations = 10^5,output= path * "/raw_data/rates_twoepoch_0.999.jld2");
```

```julia
adap = MKtest.parameters(n=50,dac=[2,4,5,10,20,30,50,70]);

analysis = filter(x-> !occursin("gz",x),readdir(path * "/raw_data/simulations/twoepochs",join=true))

true_alpha_te = DataFrame[]
abc_mk_te = DataFrame[]
grapes_te = DataFrame[]

for f in analysis
    @show f

    if occursin("0.4",f)
        h=path * "/raw_data/rates_twoepoch_0.4.jld2"        
    else
        h=path * "/raw_data/rates_twoepoch_0.999.jld2"
    end

    out = parse_inference(adap,f,h5_file=h);
    
    push!(true_alpha_te,out[1]);
    push!(abc_mk_te,out[2]);
    push!(grapes_te,out[3]);
end

true_alpha_te = vcat(true_alpha_te...);
abc_mk_te = vcat(abc_mk_te...);
grapes_te = vcat(grapes_te...);
```

```julia
CSV.write(path * "/results/abc/twoepochs/map_abcmk.tsv",abc_mk_te,delim='\t');
CSV.write(path * "/results/abc/twoepochs/true_alpha.tsv",true_alpha_te,delim='\t')
CSV.write(path * "/results/abc/twoepochs/grapes_twoepochs.tsv",true_alpha_te,delim='\t');
```

```julia
df_tgp = DataFrame[]
for i ∈ tgp_vips
    
    j = replace(i,".txt"=>"")
    folder = path * "raw_data/tgp/" * j
    file = path * "raw_data/tgp/" * i
    
    @show j
    
    mkpath(folder)
    
    out = @suppress begin
        alpha,sfs,divergence = MKtest.parse_sfs(adap,data=path * "/raw_data/tgp/tgp.txt",gene_list= file)

        summstat = MKtest.summary_statistics(adap,sfs=sfs,divergence=divergence,h5_file=path * "raw_data/rates.jld2",analysis_folder= folder,summstat_size=10^5);

        posteriors = MKtest.ABCreg(analysis_folder=folder,P=5,S=length(adap.dac),tol=0.025,abcreg=reg);
        out = MKtest.summary_abc(posteriors)
    end
    insertcols!(out[2],1,:analysis=>j)
    push!(df_tgp,out[2])
end
df = vcat(df_tgp...);
```
