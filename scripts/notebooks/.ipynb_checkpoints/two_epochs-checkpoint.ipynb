```julia
@everywhere using MKtest, RCall
@everywhere function plot_map(;analysis_folder::String,weak::Bool=true,title::String="Posteriors")

	@everywhere R"""library(locfit);library(ggplot2);library(data.table)""";

	out = filter(x -> occursin("post",x), readdir(analysis_folder,join=true))
	out          = filter(x -> !occursin(".1.",x),out)

	open(x)      = Array(CSV.read(x,DataFrame))

	# Control outlier inference. 2Nes non negative values
	flt(x)       = x[(x[:,4] .> 0) .& (x[:,1] .> 0) .& (x[:,2] .> 0) .& (x[:,3] .> 0) ,:]
	posteriors   = flt.(open.(out))

	R"""getmap <- function(df){
			temp = as.data.frame(df)
			d <-locfit(~temp[,1],temp);
			map<-temp[,1][which.max(predict(d,newdata=temp))]
		}"""

	@everywhere getmap(x)    = rcopy(R"""suppressWarnings(matrix(apply($x,2,getmap),nrow=1))""")
	
	if !weak
		posteriors = [posteriors[i][:,3:end] for i in eachindex(posteriors)]
		tmp          = progress_pmap(getmap,posteriors)
		maxp         = DataFrame(vcat(tmp...),[:α,:gam_neg,:shape])
		al           = maxp[:,1:1]
		gam          = maxp[:,2:end]
	else
		tmp          = progress_pmap(getmap,posteriors)
		maxp         = DataFrame(vcat(tmp...),[:α_weak,:α_strong,:α,:gam_neg,:shape])
		al           = maxp[:,1:3]
		gam          = maxp[:,4:end]
	end

	CSV.write(analysis_folder * "/map.tsv",maxp,delim='\t',header=true)

    
	@rput al
	@rput title
	@rput posteriors
	@rput analysis_folder

	R"""al = as.data.table(al)
			lbls = if(ncol(al) > 1){c(expression(paste('Posterior ',alpha[w])), expression(paste('Posterior ',alpha[s])),expression(paste('Posterior ',alpha)))}else{c(expression(paste('Posterior ',alpha)))}
			clrs = if(ncol(al) > 1){c('#30504f', '#e2bd9a', '#ab2710')}else{c('#ab2710')}

			if(nrow(al) == 1){
				tmp = as.data.table(posteriors[1])
				dal = suppressWarnings(data.table::melt(tmp[,1:3]))
				pal = ggplot(dal) + geom_density(aes(x=value,fill=variable),alpha=0.75) + scale_fill_manual('Posterior distribution',values=clrs ,labels=lbls) + theme_bw() + ggtitle(title) + xlab(expression(alpha)) + ylab("")
			}else{
				dal = suppressWarnings(data.table::melt(al))
				pal = suppressWarnings(ggplot(dal) + geom_density(aes(x=value,fill=variable),alpha=0.75) + scale_fill_manual('Posterior distribution',values=clrs ,labels=lbls) + theme_bw() + ggtitle(title) + xlab(expression(alpha)) + ylab(""))
			}
			suppressWarnings(ggsave(pal,filename=paste0(analysis_folder,'/map.png'),dpi=600))
			"""
	return(posteriors,maxp)
end
```

```julia
adap = MKtest.parameters(N=1000,n=50,dac=[2,4,5,10,20,25,30,50,70]);
```

```julia
"""
Executing ABC-MK
"""
function parse_inference(folder;h5_file::String="/home/jmurga/test/simulations/t_1000/reduce.jld2",bins::Union{Nothing,Int64}=nothing)

	# mkpath(output);
    analysis = split(folder,"/")[end]
    
	divergence = CSV.read.(filter(x->occursin("div",x),readdir(folder,join=true)),DataFrame,header=false);
    alpha = vcat(divergence...)
    alpha = DataFrame([alpha[:,3]./alpha[:,1],alpha[:,4]./alpha[:,1],sum.(eachrow(alpha[:,3:4]))./alpha[:,1],fill(analysis,length(divergence)),fill("slim",length(divergence))],[:α_weak,:α_strong,:α,:analysis,:method])
    divergence = map(x->Array(x)[:,1:2],divergence)
    
    sfs = CSV.read.(filter(x->occursin("sfs",x),readdir(folder,join=true)),DataFrame,header=false);
    sfs = map(x->Array(x)[:,1:3],sfs)
    
	if !isnothing(bins)
		sfs = MKtest.reduce_sfs.(sfs,bins)
	end

	summstat = MKtest.summary_statistics(param=adap,sfs=sfs,divergence=divergence,h5_file=h5_file,analysis_folder= folder,summstat_size=10^5);

	MKtest.ABCreg(analysis_folder=folder,P=5,S=length(adap.dac),tol=0.025,abcreg="/home/jmurga/ABCreg/src/reg");

	out = plot_map(analysis_folder=folder);

	insertcols!(out[2],:analysis=>analysis);
	insertcols!(out[2],:method=>"abc_mk");

	return([alpha,out[2]])
end
```
