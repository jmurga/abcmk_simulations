# No demography

```julia
adap = Analytical.parameters(N=1000,n=500,gamNeg=-457,gL=10,gH=500,dac = [2,4,5,10,20,50,200,500,700])

convolutedSamples = Analytical.binomialDict()
Analytical.binomOp!(adap,convolutedSamples.bn);
```

```julia
analysis = ["noDemog_0.4_0.1_0.2", "noDemog_0.4_0.2_0.2", "noDemog_0.4_0.3_0.2", "noDemog_0.4_0.1_0.4", "noDemog_0.4_0.2_0.4", "noDemog_0.4_0.3_0.4", "noDemog_0.4_0.1_0.8", "noDemog_0.4_0.2_0.8", "noDemog_0.4_0.3_0.8", "noDemog_0.4_0.1_0.999",  "noDemog_0.4_0.2_0.999" , "noDemog_0.4_0.3_0.999"]
n = size(analysis,1)
```

## Rates

```julia
@time df = Analytical.ratesToStats(param = adap,convolutedSamples=convolutedSamples,gH=collect(300:600),gL=collect(5:10),gamNeg=collect(-1000:-200),iterations = 10^6,shape=adap.al,output="/home/jmurga/rates.jld2");
```

## Summary statistics

```julia
h5file = jldopen("/home/jmurga/rates.jld2","r")
```

```julia
map(abcSummaries,analysis,fill(1000,n),fill(500,n),fill(h5file,n),fill(10^5,n),fill(100,n),fill("/home/jmurga/mkt/202004/rawData/summStat/noDemog_v1/",n),fill(false,n))
```

## ABC inference

```julia
inputFolder = @. "/home/jmurga/mkt/202004/rawData/summStat/noDemog_v1/" * analysis;
```

```julia
map(abcInference, inputFolder,fill(100,n),fill(5,n),fill(9,n),fill(0.01,n),fill(20,n),fill("true",n))
```

```julia
plotMap(inputFolder,1000)
```
